```python
import pandas as pd
```

```python
url="https://raw.githubusercontent.com/malindiruprabha/-Titanic-Dataset/refs/heads/main/titanic_500_dataset.csv"
```

```python
titanic_data = pd.read_csv(url)
```

```python
titanic_data.head()
```

```python
titanic_data.head(10)
```

Data Cleaning: Fixing the Titanic Data Before It Sinks Our Analysis

```python
titanic_data.isna()
```

```python
titanic_data.isna().sum()
```

```python
titanic_data = titanic_data.drop(['Cabin', 'Ticket'], axis=1, errors='ignore')
```

```python
titanic_data['Age'].median()
```

```python
titanic_data.head()
```

```python
titanic_data['Age']. fillna(titanic_data['Age'].median(), inplace=True)
```

```python
titanic_data['Age'].median()
```

```python
titanic_data['Embarked'].fillna('S', inplace=True)
```

```python
titanic_data.info()
```

```python
titanic_data.dtypes
```

Analyzing Survival Rates by Passenger class(Pclass)

```python
titanic_data.groupby ('Pclass') ['Survived'].mean()
```

Make It Attractive (Visual Thinking)

```python
import matplotlib .pyplot as plt
```

```python
Survival_Rates_by_Passenger_class=titanic_data.groupby('Pclass') ['Survived'].mean()
Survival_Rates_by_Passenger_class.plot(
    kind='bar',
    xlabel='Passenger Class',
    ylabel='Survival Rate',
    title='Survival Rates by Passenger Class',
    color =['blue', 'orange', 'green']
)
```

```python
Survival_Rates_by_Passenger_class.plot(
    kind='pie',
    color = ['#3291B6','#3291B6',"#9B5555"]
   
)
```

Analyzing Survival Rates by Gender(sex)

```python
Survival_Rates_by_Gender = titanic_data.groupby('Sex') ['Survived'].mean()
Survival_Rates_by_Gender.plot(
    kind='bar',
    xlabel='Gender',
    ylabel='Survival Rate',     
    title='Survival Rates by Gender',
    color = ['gray', 'red']
)
```

Analyzing Survival Rates by Age

```python
titanic_data.describe()
```

```python
titanic_data['AgeGroup'] = pd.cut(titanic_data['Age'],
                                  bins=[0, 2, 12, 18, 50, 100],
                                  labels=['infant', 'child', 'teen', 'adult', 'old'],
                                 
                                  )
```

```python
titanic_data.groupby('AgeGroup') ['Survived'].mean()
```

```python
# compute mean survival rate per age group and plot
Survival_Rates_by_Age = titanic_data.groupby('AgeGroup')['Survived'].mean()
Survival_Rates_by_Age.plot(
    kind='bar',
    xlabel='Age Group',
    ylabel='Survival Rate',
    title='Survival Rates by Age Group',
    color=['purple', 'cyan', 'orange', 'green', 'brown']
)
```
